```python
'''
结合 numpy 创建一个 (5, 15) 区间内，形状为 5 行 3 列的随机整数数组为 data2，并以此为基础创建 index 为 ['a', 'c', 'e', 'f', 'h']，
columns 为 ['one', 'two', 'three'] 的 DataFrame，并命名为 df_test（阅读参考文章④）
'''

data2 = np.random.randint(5,15,15).reshape(5,3)
df_test = pd.DataFrame(data2,index = ['a', 'c', 'e', 'f', 'h'],columns = ['one', 'two', 'three'])
df_test
```

```python
'''
运用 loc 索引以及替换的知识，按下列要求对 df_test 进行增、改、替操作。
完成上述要求后，将该数据命名为 df_change（任务提示：一般用 df.loc[行，列] = ？ 做替换。阅读参考文章⑤⑥）
(1) 将 a 行 one 列处替换成空值;(2) 将 c 行 two 列处替换成 -99;(3) 将 c 行 three 列处替换成 -99;(4) 将 a 行 two 列处替换成 -100;
(5) 新增 four 列，值为 test;(6) 重置索引为 ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
https://blog.csdn.net/The_Time_Runner/article/details/84075409
https://blog.csdn.net/youngbit007/article/details/53067707
'''

# DataFrame如何对某行某列的单个元素进行替换(loc[]、iloc[]、at[]区别)
# 在DataFrame文件中，对i行j列的那个元素进行替换、插入数据操作，采用的是pd.loc[]
# pd.iloc[i,j] i,j 只能是数字，即原始索引。
# pd.loc[i,‘j’] i,j既可以是数字，也可是字母，即索引标签；更适用于一段数据。
# pd.at[i,‘j’] i,j既可以是数字，也可是字母，不同的是，当定位的是某个具体的元素时，用at[]效率更高。
# 这两个函数经常用来提取指定位置的数据，直接对其赋值也是可以的。
# pd.loc[i,j] = value

# pandas的reindex对象，是数据符合新的索引来构造一个新的对象
# Series的reindex使它符合新的索引，如果索引的值不存在就填入缺失值

# method选项来控制填充值或内插值：
# method : {None, ‘backfill’/’bfill’, ‘pad’/’ffill’, ‘nearest’}, optional。
# ffill/pad 向前或进位填充，bfill/backfill 向后或补位填充

# (1) 将 a 行 one 列处替换成空值
df_test.loc['a','one'] = ''

# df_test.loc['c','two':'three']=-99
# (2) 将 c 行 two 列处替换成 -99
df_test.loc['c','two'] = -99
# (3) 将 c 行 three 列处替换成 -99
df_test.loc['c','three'] = -99

# (4) 将 a 行 two 列处替换成 -100
df_test.loc['a','two'] = -100

# (5) 新增 four 列，值为 test
df_test.loc[:,'four'] = 'test'

# (6) 重置索引为 ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
df_change = df_test.reindex(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
```
