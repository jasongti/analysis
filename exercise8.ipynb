#### 请统计出不同户型对应的房源数量、平均面积；并从大到小进行排序，判断出主流户型是什么？平均面积多大？（任务提示：以 huxing 为分组对象，展示 size 列，统计户型数量和 size 平均数。阅读参考文章⑤⑥⑦⑧）

```python
'''
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.agg.html
'''

#不同户型的房源数量
# data_new.huxing.value_counts()
#不同户型的平均面积
# data_new['size'].groupby(data_new.huxing).mean().sort_values()
data_cat=data_new.groupby('huxing')['size'].agg([('房源数量','count'),('平均面积','mean')]).sort_values(by=['房源数量','平均面积'],ascending=[False,True])
data_cat
```

```python
#主流户型与平均面积
data_cat.head(10)
```

#### 获取每个小区的房源数量，并按从大到小的顺序排序（阅读参考文章⑤⑥⑦⑧）

```python
data_new.groupby('xiaoqu_name')['xiaoqu_name'].agg([('房源数量','count')]).sort_values(by='房源数量',ascending=False)
```

#### 请统计出上海每个小区每平方米的均价（阅读参考文章⑤⑥⑦⑧）

```python
data_new.groupby('xiaoqu_name')['up_price'].agg([('均价','mean')]).sort_values(by='均价',ascending=False)
```

#### 请计算出不同年代的房子数量（10年一个区间）（任务提示：先将 buildyear 列转为时间序列，再利用 groupby 函数统计出每年的房源数量，最后用 resample 函数进行汇总。阅读参考文章⑨⑩⑪）
>⑨　在DataFrame中如何将字符串转换为float（也可以参考转成str） https://cloud.tencent.com/developer/ask/47544

>⑩　python中时间转换datetime和pd.to_datetime详析 http://www.3qphp.com/python/pybase/4606.html

>⑪　resample按时间聚合 https://blog.csdn.net/sinat_41701878/article/details/80491631

```python
data_new.buildyear = data_new.buildyear.astype('str')
data_new.buildyear = pd.to_datetime(data_new.buildyear)
data_new
```

```python
data_by_year = data_new.groupby('buildyear')['_id'].agg([('房源数量','count')])
data_by_year
```

```python
data_by_year.resample('10y').sum()
```
